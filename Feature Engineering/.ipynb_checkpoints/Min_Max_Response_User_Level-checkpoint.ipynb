```python
import pandas as pd
import numpy as np
```

```python
#read in the data
votes=pd.read_csv("C:/Users/Ergo/Desktop/Git_Dotin/Dotin-Columbia-Castone-Team-Alpha-/Data/Clean Data/votes_v2.csv")
```

```python
votes.head()
```

```python
def get_min_response(df):
    """Create new column to capture absolute min response (aka 1) as 1=yes, 0=no"""
    if df['score']==1:
        return 1
    else:
        return 0

votes['is_min_response?']= votes.apply(lambda x: get_min_response(x), axis=1)
```

```python
def get_max_response(df):
    """Create new column to capture absolute max response (aka 5 or 7 depending on question) as 1=yes, 0=no"""
    if "pgi_" in df['value']:
        if df['score']==7:
            return 1
        else:
            return 0
    else:
        if df['score']==5:
            return 1
        else: 
            return 0
        
votes["is_max_response?"]=votes.apply(lambda x: get_max_response(x), axis=1)
```

```python
#function to group survey responses by type and see if they have been responded as an absolute min and max 
#as an aggregate

def create_question_buckets(df):
    """Create a new column to bucket questions by category"""
    if "bf_" in df['value']:
        return "bf_questions"
    elif "bs_" in df['value']:
        return "bs_questions"
    elif "miq_" in df['value']:
        return "miq_questions"
    else:
        return "pgi_questions"

votes['question_category']=votes.apply(lambda x: create_question_buckets(x), axis=1)
```

```python
#ABSOLUTE MIN/MAX RESPONSE CALCULATIONS#

#create pivot table for absolute min/max answer to aggregate information

abs_minmax_pivot_table=pd.pivot_table(votes, index=['user_id'], columns=['question_category'],values=['is_min_response?', 
                                'is_max_response?'],aggfunc=[np.sum, len])
```

```python
#turn pivot table to regular dataframe
abs_minmax_pivot_table.columns = abs_minmax_pivot_table.columns.droplevel(0)
abs_minmax_votes = abs_minmax_pivot_table.reset_index().rename_axis(None, axis=1)
```

```python
#change column names for is_min_responses
abs_minmax_votes.columns=['user_id','is_max_bf_questions','is_max_bs_questions','is_max_miq_questions','is_max_pgi_questions',
                          'is_min_bf_questions', 'is_min_bs_questions','is_min_miq_questions','is_min_pgi_questions',
                         'bf_max_len','bs_max_len','miq_max_len','pgi_max_len',
                         'bf_min_len','bs_min_len','miq_min_len', 'pgi_min_len']
```

```python
abs_minmax_votes.head()
```

```python
#function to capture users who have selected only absolute min or max per question category

#for bf_questions
def abs_min_users_bf(df):
    """Create new column to capture users who have full absolute min response for bf_questions"""
    
    if df['is_min_bf_questions']==df['bf_max_len']:
        return 1
    else:
        return 0

abs_minmax_votes['bf_abs_min_response?']=abs_minmax_votes.apply(lambda x: abs_min_users_bf(x), axis=1)

################################
def abs_max_users_bf(df):
    """Create new column to capture users who have full absolute min response for bf_questions"""
    
    if df['is_max_bf_questions']==df['bf_max_len']:
        return 1
    else:
        return 0
    
abs_minmax_votes['bf_abs_max_response?']=abs_minmax_votes.apply(lambda x: abs_max_users_bf(x), axis=1)
```

```python
#function to capture users who have selected only absolute min or max per question category

#for miq_questions
def abs_min_users_miq(df):
    """Create new column to capture users who have full absolute min response for miq questions"""
    
    if df['is_min_miq_questions']==df['miq_max_len']:
        return 1
    else:
        return 0

abs_minmax_votes['miq_abs_min_response?']=abs_minmax_votes.apply(lambda x: abs_min_users_miq(x), axis=1)

################################
def abs_max_users_miq(df):
    """Create new column to capture users who have full absolute min response for miq questions"""
    
    if df['is_max_miq_questions']==df['miq_max_len']:
        return 1
    else:
        return 0
    
abs_minmax_votes['miq_abs_max_response?']=abs_minmax_votes.apply(lambda x: abs_max_users_miq(x), axis=1)
```

```python
#extract only the question-abs-minmax-response? features
final_dataset=abs_minmax_votes[['user_id','bf_abs_min_response?','bf_abs_max_response?',
                               'bs_abs_min_response?','bs_abs_max_response?',
                               'miq_abs_min_response?', 'miq_abs_max_response?',
                               'pgi_abs_min_response?','pgi_abs_max_response?']]
```
